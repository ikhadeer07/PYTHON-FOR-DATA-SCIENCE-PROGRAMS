```python
import pandas as pd
import matplotlib.pyplot as plt
# Import CSV file to DataFrame
df = pd.read_csv('/content/student.csv')
# Handle missing data
df.dropna()  # Drop rows with missing values
df.fillna(0)  # Fill missing values with 0
print(df)
# Transform data using apply() and map()
df['Age'] = df['Age'].apply(lambda x: x * 2)  # Apply a function to a column
#df['Age'] = df['Column2'].map({'A': 1, 'B': 2, 'C': 3})  # Map values in a column
# Detect and filter outliers
outliers = df[(df['Age'] > 100) | (df['Age'] < 10)]  # Detect outliers based on a conditio
filtered_df = df[(df['Age'] >= 10) & (df['Age'] <= 100)]  # Filter out outliers
print(outliers)
print(filtered_df)
# Perform vectorized string operations
df['city'] = df['city'].str.upper()  # Convert strings to uppercase
print(df)
# Visualize data
df.plot(kind='line', x='Age', y='weight')  # Line plot
df.plot(kind='bar', x='Age', y='Height')  # Bar plot
x=df['Age'].plot(kind='hist')  # Histogram
df['Age'].plot(kind='density')  # Density plot
df.plot(kind='scatter', x='Age', y='marks')  # Scatter plot
plt.show()  # Show the plots
plt.show(x)
```
