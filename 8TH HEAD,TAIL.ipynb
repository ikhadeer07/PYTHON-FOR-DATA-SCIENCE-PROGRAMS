```python
import pandas as pd
# Import CSV file to DataFrame
df = pd.read_csv('/content/student.csv')
# Visualize first and last 10 records
print("First 10 records:")
print(df.head(10))
print()
print("Last 10 records:")
print(df.tail(10))
print()
# Get shape, index, and column details
print("Shape of DataFrame:", df.shape)
print("Index of DataFrame:")
print(df.index)
print("Columns of DataFrame:")
print(df.columns)
print()
# Select/Delete records (rows) based on conditions
selected_rows = df[df['marks'] > 50]  # Select rows where Column1 > 50
deleted_rows = df[df['Age'] != 20]  # Delete rows where Column2 is not 'A'
# Select/Delete columns based on conditions
selected_columns = df[['Age', 'Height']]  # Select specific columns
deleted_columns = df.drop(['weight', 'city'], axis=1)  # Delete specific columns
# Perform ranking and sorting operations
df['Rank'] = df['weight'].rank(ascending=False)  # Add a new column with ranking based on
sorted_df = df.sort_values(by='Age', ascending=True)  # Sort DataFrame by Column1
# Statistical operations on columns
mean_column1 = df['Age'].mean()  # Mean of Column1
median_column2 = df['weight'].median()  # Median of Column2
std_column3 = df['Height'].std()  # Standard deviation of Column3
# Count and uniqueness of categorical values
categorical_counts = df['Name'].value_counts()  # Count of each category
unique_values = df['Age'].unique()  # Unique values in Column4
# Rename single/multiple columns
df = df.rename(columns={'weight': 'NewColumn1', 'City': 'NewColumn2'})
# Print the modified DataFrame and other results
print("Selected Rows:")
print(selected_rows)
print()
print("Deleted Rows:")
print(deleted_rows)
print()
print("Selected Columns:")
print(selected_columns)
print()
print("Deleted Columns:")
print(deleted_columns)
print()
print("Ranking and Sorting:")
print(df)
print()
print("Mean of Column1:", mean_column1)
print("Median of Column2:", median_column2)
print("Standard Deviation of Column3:", std_column3)
print()
print("Categorical Value Counts:")
print(categorical_counts)
print()
print("Unique Values in Column4:")
print(unique_values)
df.loc[df['Name'] == 'mahammad', 'Name'] = 'muskan'
print(df)
```
